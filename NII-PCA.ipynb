### Convert a single image into FBX without using PCA

```python
import fbx
from skimage import measure
import nibabel as nib
from scipy.ndimage import gaussian_filter

# Load the NIfTI file
voxel_data = images[0].astype(int)
# voxel_data = transformed_img_array

# Can choose to apply gaussian filter
# smoothed_data = gaussian_filter(voxel_data, sigma=1)  # Adjust sigma for smoothing intensity
# verts, faces, normals, _ = measure.marching_cubes(smoothed_data, level=0.3)
verts, faces, normals, _ = measure.marching_cubes(voxel_data, level=0.3)
print (f"Original model has {len(verts)} vertices and {len(faces)} of faces")

# Initialize the FBX SDK
manager = fbx.FbxManager.Create()
scene = fbx.FbxScene.Create(manager, "MyScene")

# Create a mesh object
mesh = fbx.FbxMesh.Create(scene, "MyMesh")

# Set the vertices in the FBX mesh
mesh.InitControlPoints(len(verts))
for i, vert in enumerate(verts):
    mesh.SetControlPointAt(fbx.FbxVector4(*vert), i)

# Add polygons to the mesh
for face in faces:
    mesh.BeginPolygon()
    for idx in face:
        mesh.AddPolygon(idx)
    mesh.EndPolygon()

# Create a node for the mesh
mesh_node = fbx.FbxNode.Create(scene, "MyMeshNode")
mesh_node.SetNodeAttribute(mesh)

# Add the mesh node to the scene
scene.GetRootNode().AddChild(mesh_node)

# Export the scene to an FBX file
exporter = fbx.FbxExporter.Create(manager, "")
exporter.Initialize("Single fbx without PCA.fbx", -1, manager.GetIOSettings())
exporter.Export(scene)
exporter.Destroy()
manager.Destroy()
print("Single Warped FBX")


import os

# Specify the path to your FBX file
fbx_file_path = "Single fbx without PCA.fbx"

# Check if the file exists and get its size in MB
if os.path.exists(fbx_file_path):
    file_size_bytes = os.path.getsize(fbx_file_path)
    file_size_mb = file_size_bytes / (1024 * 1024)  # Convert bytes to MB
    print(f"File size: {file_size_mb:.2f} MB")
else:
    print("File not found")
```

### PCA Function

```python
from scipy.sparse.linalg import eigsh

def pca(data, n_comps):
    x_hat = data - np.mean(data, axis=1, keepdims=True)
    cov = np.cov(x_hat, rowvar=True)
    eigenvalues, eigenvectors = eigsh(cov, k=n_comps)
    return eigenvalues, eigenvectors
```

### Sparsity Check Funcion

```python
def calculate_sparsity_3d(matrix):
    if not isinstance(matrix, list) or not all(isinstance(i, list) for i in matrix):
        raise ValueError("Input must be a 3D Python list.")
    
    total_elements = 0
    zero_elements = 0
    
    # Iterate through the 3D list
    for layer in matrix:
        for row in layer:
            for element in row:
                total_elements += 1
                if element == 0:
                    zero_elements += 1
    
    # Calculate sparsity
    if total_elements == 0:
        raise ValueError("The input matrix is empty.")
    
    sparsity = zero_elements / total_elements
    return sparsity

```

```python
from sklearn.decomposition import PCA

image_flat = images[0].reshape(images[0].shape[0], -1)

num_comps = [10, 20, 50]

reconstructed_images_PCA = []

for n in num_comps:
    print(f"\nReconstructing with {n} components...")
    # Perform PCA on the flattened first image
    pca = PCA(n_components=n)
    transformed = pca.fit_transform(image_flat)  # Reduce dimensions
    reconstructed_flat = pca.inverse_transform(transformed)  # Reconstruct image

    # Reshape back to the original 3D shape
    reconstructed_img = reconstructed_flat.reshape(images[0].shape)
    reconstructed_images_PCA.append(reconstructed_img)

# Sanity check the shapes of reconstructed images
for idx, n in enumerate(num_comps):
    print(f"Reconstructed image with {n} components has shape: {reconstructed_images_PCA[idx].shape}")
```

```python
from sklearn.decomposition import SparsePCA

# Flatten the first image (as in your original code)
image_flat = images[0].reshape(images[0].shape[0], -1)

# Define the number of components to test
num_comps = [100, 200]

# Initialize list for reconstructed images
reconstructed_images_SPCA = []

# Loop through each number of components
for n in num_comps:
    print(f"\nReconstructing with {n} components using Sparse PCA...")
    
    # Perform Sparse PCA on the flattened image
    spca = SparsePCA(n_components=n, alpha=1, random_state=0)
    transformed = spca.fit_transform(image_flat)  # Reduce dimensions
    reconstructed_flat = transformed @ spca.components_  # Reconstruct image
    print("here")
    # Reshape back to the original 3D shape
    reconstructed_img = reconstructed_flat.reshape(images[0].shape)
    reconstructed_images_SPCA.append(reconstructed_img)

# Sanity check the shapes of reconstructed images
for idx, n in enumerate(num_comps):
    print(f"Reconstructed image with {n} components has shape: {reconstructed_images_SPCA[idx].shape}")
```

### Convert PCA Reconstruction to FBX

```python
import fbx
from skimage import measure
import os

# Loop over the reconstructed images and save them as FBX
for i, reconstructed_image in enumerate(reconstructed_images_PCA):
    num_comp = num_comps[i]  # Get the corresponding number of components

    # Prepare voxel data for marching cubes
    voxel_data = reconstructed_image.astype(int)

    # Apply marching cubes to create vertices and faces
    verts, faces, normals, _ = measure.marching_cubes(voxel_data, level=0.3)
    print(f"Reconstructed {num_comp} has {len(verts)} vertices and {len(faces)} faces")

    # Initialize the FBX SDK
    manager = fbx.FbxManager.Create()
    scene = fbx.FbxScene.Create(manager, "MyScene")

    # Create a mesh object
    mesh = fbx.FbxMesh.Create(scene, "MyMesh")

    # Set the vertices in the FBX mesh
    mesh.InitControlPoints(len(verts))
    for j, vert in enumerate(verts):
        mesh.SetControlPointAt(fbx.FbxVector4(*vert), j)

    # Add polygons to the mesh
    for face in faces:
        mesh.BeginPolygon()
        for idx in face:
            mesh.AddPolygon(idx)
        mesh.EndPolygon()

    # Create a node for the mesh
    mesh_node = fbx.FbxNode.Create(scene, "MyMeshNode")
    mesh_node.SetNodeAttribute(mesh)

    # Add the mesh node to the scene
    scene.GetRootNode().AddChild(mesh_node)

    # Export the scene to an FBX file
    fbx_file_name = f"./PCA Reconstructed Models/PCA_reconstructed_{num_comp}.fbx"
    exporter = fbx.FbxExporter.Create(manager, "")
    exporter.Initialize(fbx_file_name, -1, manager.GetIOSettings())
    exporter.Export(scene)
    exporter.Destroy()
    manager.Destroy()

    # Print success message
    print(f"FBX file created: {fbx_file_name}")

    # Check if the file exists and get its size in MB
    if os.path.exists(fbx_file_name):
        file_size_bytes = os.path.getsize(fbx_file_name)
        file_size_mb = file_size_bytes / (1024 * 1024)  # Convert bytes to MB
        print(f"File size of {fbx_file_name}: {file_size_mb:.2f} MB")
    else:
        print(f"File {fbx_file_name} not found")
```
